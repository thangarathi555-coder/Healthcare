```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)
```

```python
n = 10000
```

```python
data = pd.DataFrame({
    "Patient_ID": range(1, n+1),
    "Age": np.random.randint(20, 81, n),
    "Gender": np.random.choice(["Male", "Female"], n),
    "Blood_Pressure": np.random.randint(80, 181, n),
    "Sugar_Level": np.random.randint(70, 201, n),
    "Cholesterol": np.random.randint(150, 301, n),
    "Heart_Rate": np.random.randint(60, 121, n)
})
```

```python
data.to_csv("healthcare_data.csv", index=False)
```

```python
print("Dataset Created Successfully!\n")
print(data.head())
```

```python
print("\n--- Data Info ---")
print(data.info())
```

```python

print("\n--- Missing Values ---")
print(data.isnull().sum())
```

```python
data = data.dropna()
```

```python
print("\n--- Statistical Summary ---")
print(data.describe())
```

```python
def risk_level(row):
    if (row["Blood_Pressure"] > 140 or
        row["Sugar_Level"] > 160 or
        row["Cholesterol"] > 240):
        return "High"
    elif (row["Blood_Pressure"] > 120 or
          row["Sugar_Level"] > 120 or
          row["Cholesterol"] > 200):
        return "Medium"
    else:
        return "Low"
```

```python
data["Risk"] = data.apply(risk_level, axis=1)
display(data.head())
```

```python
high_risk = data[data["Risk"] == "High"]
```

```python
print("\nTotal High Risk Patients:", len(high_risk))
print(high_risk.head())
```

```python
plt.hist(data["Age"])
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
data["Risk"].value_counts().plot(kind='bar')
plt.title("Risk Level Distribution")
plt.show()
```

```python
plt.scatter(data["Age"], data["Blood_Pressure"])
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.title("Age vs Blood Pressure")
plt.show()
```

```python
sns.boxplot(x=data["Cholesterol"])
plt.title("Cholesterol Outliers")
plt.show()
```

```python
corr = data.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
print("\n--- Gender vs Risk Level ---")
print(data.groupby("Gender")["Risk"].value_counts())
```

```python

print("\n--- Age vs Average Blood Pressure ---")
print(data.groupby("Age")["Blood_Pressure"].mean())
```
